#### 예시 : 세 가지 수를 전달받아 가장 큰 값을 반환하는 함수

```python
def big(x,y,z):
    big_n = x
    
    if x > y :
        if x > z :
            big_n = x
        else :
            big_n = z
    else:
        if y > z :
            big_n = y
        else : 
            big_n = z
            
    return big_n
```

```python
a = big(x=10, y=15, z=5)
print(a)
```

#### *args 인자는 매개변수 수가 불확실할 때 사용하는 방법으로 사용자가 나중에 넣는 만큼 받아들임
#### 예를 들어, 자유롭게 파라미터 수를 *args 로 받아서 모두 반환할 때는 return 에서 * 을 뺀 args로 반환할 수 있음

```python
def parameter(*args):
    return args
```

```python
print(parameter(1,2,3,4,5))
```

```python
print(parameter(1,2))
```

```python
print(parameter(1,3,5))
```

#### 
#### 예를 들어, 몇 개인지 모르는 점수를 입력받아서 평균을 내주는 함수를 만들어보면

```python
def average_cal(*args):
    length = len(args)
    
    total = 0
    for i in args:
        total = total + i
    
    avg = total / length
    return avg
```

```python
avg = average_cal(10, 20, 30)
print(avg)
```

```python
avg = average_cal(10, 20, 30, 50 ,100)
print(avg)
```

## 전역변수와 지역변수

#### 프로그래밍에서 변수는 크게 전역변수와 지역변수로 구분할 수 있음
### 전역변수 : 코드 전체에서 사용되는 변수
### 지역변수 : 특정 범위 안에서만 사용되는 변수

#### 그림의 예로는, 독수리는 전역변수, 돼지와 닭은 지역변수로 해석할 수 있음

### 


```python
a = 20

def f1():
    a = 10
    print("지역변수 : ",a)

print("전역변수 : ",a)

f1()
```

### 
#### 만약 여러 개 함수가 똑같은 이름의 변수를 사용한다하더라도 각 함수 안에서의 지역변수들임

```python
a = 20

def f1():
    a = 10
    print(a)

def f2():
    a = 5
    print(a)
    
def f3():
    a = 1
    print(a)
    
print(a)
f1()
f2()
f3()
```

#### 
#### 함수에서 선언한 지역변수를 함수를 벗어난 다른 곳에서는 사용할 수 없음

```python
def f1():
    q = 10
    return q

print(q)
```

### 
### global : 변수를 전역변수로 바꾸는 함수
### local : 변수를 지역변수로 바꾸는 함수

```python
a = 10

def func():
    global a  # a 변수를 전역 변수로 선언
    a = 20    # 전역 변수 a의 값을 변경
    print(a)

func()  # 출력: 20
print(a)  # 출력: 20 (함수 내에서 변경된 전역 변수의 값)
```

#### 문제 : 로또 번호를 하나씩 생성하는 함수를 반복문을 사용하여 구현해보세요, 단 로또 번호가 중복되지 않아야 합니다.

#### 아래 random 이라는 난수를 생성하는 라이브러리에 random.randint(a, b) 함수는 a에서 b까지 숫자 중 임의로 하나를 반환하는 함수임

```python
import random
lotto_num = random.randint(1,45)
print(lotto_num)
```

```python
def gen_lotto():
    n = random.randint(1, 45)
    return n

print("로또 추첨을 시작합니다.")
lotto_list = []

while(True):
    
    num = gen_lotto()
    
    if num in lotto_list :
        continue
    else :
        lotto_list.append(num)
    
    if len(lotto_list) > 6:
        break

print("생성된 로또 번호 : ", lotto_list)
```
