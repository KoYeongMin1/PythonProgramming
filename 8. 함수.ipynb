```python
def big(x,y,z):
    big_n = x
    
    if x > y :
        if x > z :
            big_n = x
        else :
            big_n = z
    else:
        if y > z :
            big_n = y
        else : 
            big_n = z
            
    return big_n
```

```python
a = big(x=10, y=15, z=5)
print(a)
```

####  ,         

```python
def average_cal(*args):
    length = len(args)
    
    total = 0
    for i in args:
        total = total + i
    
    avg = total / length
    return avg
```

```python
avg = average_cal(10, 20, 30)
print(avg)
```

```python
avg = average_cal(10, 20, 30, 50 ,100)
print(avg)
```

####  random     random.randint(a, b)  a b      

```python
import random
lotto_num = random.randint(1,45)
print(lotto_num)
```

```python
def gen_lotto():
    n = random.randint(1, 45)
    return n

print("  .")
lotto_list = []

while(True):
    
    num = gen_lotto()
    
    if num in lotto_list :
        continue
    else :
        lotto_list.append(num)
    
    if len(lotto_list) > 6:
        break

print("   : ", lotto_list)
```
